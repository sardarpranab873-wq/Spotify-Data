```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


data = {
    'Energy (binned)': ['(-1.002, -0.667]', '(-0.667, -0.333]', '(-0.333, 0.0]', '(0.0, 0.333]', '(0.333, 0.667]', '(0.667, 1.0]'],
    'Average Popularity': [5.0, 0.0, 4.0, 10.7, 14.3, 17.8]
}
df = pd.DataFrame(data)


x_values = df['Energy (binned)']
y_values = df['Average Popularity']


fig_bg_color = '#fff0ed'
line_color = 'steelblue'
point_color = 'steelblue'
baseline_color = 'red'


fig, ax = plt.subplots(figsize=(10, 6))


fig.set_facecolor(fig_bg_color)
ax.set_facecolor(fig_bg_color)


for i in range(len(df)):
    if y_values[i] > 0:
        ax.plot([x_values[i], x_values[i]], [0, y_values[i]],
                color=line_color,
                linewidth=1.5,
                zorder=1)


ax.scatter(
    x=x_values,
    y=y_values,
    color=point_color,
    s=70,
    zorder=2
)


ax.axhline(y=0, color=baseline_color, linestyle='-', linewidth=2)
ax.set_title('Lollipop Chart: Average Popularity by Energy', fontsize=16)
ax.set_xlabel('Energy (binned)', fontsize=12)
ax.set_ylabel('Average Popularity', fontsize=12)


plt.xticks(rotation=30, ha='right')


plt.ylim(-0.5, 18.5)
plt.yticks(np.arange(0, 18.5, 2.5))


ax.grid(axis='y', linestyle='-', alpha=0.5, color='lightgray')
ax.grid(axis='x', visible=False)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.spines['bottom'].set_visible(False)


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)
n_points = 5000

durations = np.concatenate([
    np.random.gamma(shape=2, scale=150000, size=int(0.9 * n_points)),
    np.random.uniform(1000000, 5000000, size=int(0.1 * n_points))
])

popularity_clustered = np.clip(np.random.normal(loc=50 - (durations[:int(0.9 * n_points)] / 100000) * 10, scale=30, size=int(0.9 * n_points)), 0, 95)
popularity_outliers = np.random.uniform(0, 50, size=int(0.1 * n_points))
popularity = np.concatenate([popularity_clustered, popularity_outliers])

df = pd.DataFrame({'Duration (ms)': durations, 'Popularity': popularity}).sample(n=3000, random_state=42).reset_index(drop=True)
df = df[df['Duration (ms)'] < 5000000]


fig_bg_color = '#fff0ed'
point_color = 'coral'
grid_color = 'lightgray'


fig, ax = plt.subplots(figsize=(10, 5))


fig.set_facecolor(fig_bg_color)
ax.set_facecolor(fig_bg_color)


ax.scatter(
    x=df['Duration (ms)'],
    y=df['Popularity'],
    color=point_color,
    alpha=0.5,
    edgecolor='white',
    linewidth=0.5,
    s=20
)


ax.set_title('Duration vs Popularity', fontsize=16)
ax.set_xlabel('Duration (ms)', fontsize=12)
ax.set_ylabel('Popularity', fontsize=12)


ax.set_xticks(np.arange(0, 5000001, 1000000))
ax.set_yticks(np.arange(0, 100, 20))


ax.grid(True, linestyle='-', alpha=0.3, color=grid_color)

ax.set_axisbelow(True)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.spines['left'].set_color(grid_color)
ax.spines['bottom'].set_color(grid_color)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


years = np.arange(1970, 2025)


tracks = np.array([

    40, 45, 50, 48, 55, 60, 62, 58, 65, 70,

    75, 85, 95, 105, 120, 140, 150, 170, 180, 200,

    230, 210, 260, 300, 280, 330, 350, 320, 370, 360,

    380, 400, 420, 450, 520, 610, 700, 750, 820, 900,

    1200, 1700, 3300, 2700, 2500, 2800, 2900, 2950, 3000, 3100,

    3800, 5000, 6200, 5500, 5700
])


if years.size != tracks.size:
    raise ValueError(f"years ({years.size}) and tracks ({tracks.size}) must have the same length")


fig, ax = plt.subplots(figsize=(14, 6))


ax.set_facecolor('#f3e5e7')


ax.plot(years, tracks,
        marker='o', linestyle='-', linewidth=2, markersize=6,
        markeredgewidth=1.5, color='darkgreen', markerfacecolor='green', markeredgecolor='darkgreen')


for y in (2012, 2022, 2023):
    if y in years:
        idx = int(np.where(years == y)[0])
        ax.annotate(f'{tracks[idx]:,}', xy=(y, tracks[idx]), xytext=(0, 9),
                    textcoords='offset points', ha='center', fontsize=9, fontweight='bold')


ax.set_title('Number of Tracks Released Per Year', fontsize=18, pad=16)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Number of Tracks', fontsize=12)


ax.set_xticks(np.arange(1970, 2025, 5))
ax.set_xlim(years.min() - 1, years.max() + 1)


ymax = int(np.ceil(tracks.max() / 500.0) * 500)
ax.set_yticks(np.linspace(0, ymax, 7))
ax.set_ylim(0, ymax + 100)


ax.grid(True, which='major', linestyle='-', linewidth=0.8, alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


years = np.arange(1970, 2025)


np.random.seed(2)
danceability = 0.55 + 0.1 * np.sin((years - 1970) / 3) + np.random.normal(0, 0.03, len(years))
energy = 0.6 + 0.12 * np.cos((years - 1970) / 2.8) + np.random.normal(0, 0.05, len(years))


danceability = np.clip(danceability, 0.4, 0.75)
energy = np.clip(energy, 0.3, 0.75)

fig, ax = plt.subplots(figsize=(12, 6))


ax.set_facecolor('#e8d6d7')


ax.plot(years, danceability, marker='o', color='navy', linewidth=2,
        markersize=6, label='Danceability')
ax.plot(years, energy, marker='s', color='sienna', linewidth=2,
        markersize=6, label='Energy')


ax.set_title('Trends in Danceability and Energy of Popular Songs', fontsize=16, pad=12)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Average Value', fontsize=12)


ax.set_xlim(1970, 2025)
ax.set_ylim(0.3, 0.8)
ax.set_xticks(np.arange(1970, 2025, 10))
ax.set_yticks(np.arange(0.3, 0.81, 0.1))


legend = ax.legend(frameon=True, loc='lower right')
legend.get_frame().set_facecolor('#f0f0f0')
legend.get_frame().set_edgecolor('gray')


ax.grid(True, linestyle='-', linewidth=0.8, alpha=0.6)

plt.tight_layout()
plt.show()
```
